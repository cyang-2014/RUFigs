# Figure S10

Simulated data for sequencing synthetic 3kb amplicons.

```r
library(ggplot2)
library("gridExtra")
library("plyr")
library("rgl")

testingAMPLICONS11_2 = read.csv("data/simulations/readuntil100x_11amps_2.log", header = TRUE,sep=",")

testingAMPLICONS11_2$OddEven = "RU"
head(testingAMPLICONS11_2)

```

```r
FigS9a<-ggplot(data=subset(testingAMPLICONS11_2,read_reason!="Stopped"),aes(x=read_event_count,fill=read_reason)) + geom_histogram(binwidth=50,position="identity",alpha = 1) + theme_bw() + theme(text = element_text(size=8),legend.position="bottom") + labs(y="Read Count",x="Read Length (events)",title = "A) Template Read Lengths 11 Amplicons")+ scale_fill_discrete(guide = guide_legend(title = "Read Type"))
FigS9b<-ggplot(data=subset(testingAMPLICONS11_2,read_reason=="Unblock"),aes(x=read_event_count,y=seq_index,color=read_start)) + geom_point() + theme_bw() + theme(text = element_text(size=8))+ scale_colour_gradient(name="Time (S)") + labs(y="Channel Number",x="Read Length (events)",title = "B) Read End Times 11 Amplicons")

FigS9a
FigS9b

```
